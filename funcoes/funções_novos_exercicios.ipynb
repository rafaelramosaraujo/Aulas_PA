# 1) com retorno e sem parâmetro:

```python
def gerador_senha():

  senha = ""

  for i in range(2):
    senha = senha + random.choice(string.ascii_uppercase)
    senha = senha + random.choice(string.ascii_lowercase)
    senha = senha + random.choice(string.digits)
    senha = senha + random.choice(string.punctuation)

  return senha
```

```python
gerador_senha()
```

# 2) Crie uma função com parÂmetro e sem retorno chamada mostrar_tabuada(numero) que recebe um parâmetro inteiro e imprime na tela a tabuada de 1 a 10 desse número.


```python
def tabuada(numero):
  tabuada = []

  for i in range(1,11):
    tabuada.append(numero * i)
    i + 1

  print(tabuada)
```

# 3) Crie uma função com retorno e com parâmetro chamada validar_senha(senha) que recebe uma string e retorna um dicionário contendo:


```python
def validar_senha(senha):
  resultados = {}
  verificacao = list(senha)
  tem_maiuscula = 0
  tem_minuscula = 0
  tem_numero = 0
  tem_especial = len(senha)

  for i in range(8):
    if verificacao[i].isupper() == True:
      tem_maiuscula = tem_maiuscula + 1
    if verificacao[i].islower() == True:
      tem_minuscula = tem_minuscula + 1
    if verificacao[i].isdigit() == True:
      tem_numero = tem_numero + 1
    if verificacao[i].isalnum() == True:
      tem_especial = tem_especial - 1

  if len(senha) >= 8:
    resultados["tamanho_valido"] = "True"
    if len(senha) >= 8 and tem_maiuscula >= 2 and tem_minuscula >= 2 and tem_numero >= 2 and tem_especial <= 2:
      print("sua senha é segura e tem todos o requisitos")
    else:
      print("sua senha não é segura")
  else:
    print("sua senha não é grande o suficiente")
    resultados["tamanho_valido"] = "Falso"

  if tem_maiuscula >= 1:
    resultados["tem_maiuscula"] = "True"
  else:
    resultados["tem_maiuscula"] = "False"
  if tem_minuscula >= 1:
    resultados["tem_minuscula"] = "True"
  else:
    resultados["tem_minuscula"] = "False"
  if tem_numero >= 1:
    resultados["tem_numero"] = "True"
  else:
    resultados["tem_numero"] = "False"
  if tem_especial <= len(senha) -1 :
    resultados["tem_especial"] = "True"
  else:
    resultados["tem_especial"] = "False"

  return resultados

```
